# Imputing Missing Values
### column  "emp_length" has highest missing values i.e 2.68 % 
### Column  "revol_util" has 0.13 % missing values
### Now lets check type of data each column has

```python
loan_data.info()
```

```python
print("Mode of emp_length is : " + loan_data.emp_length.mode()[0])
loan_data.emp_length.value_counts()
```

- Above we can clearly see that the mode value of 'emp_length' i.e. "10+ years" has the highest frequency in the dataframe
- Also the missing value percentage is very low.
- So we can easily impute mode value in missing values & it will not have any impact on our analysis.

```python
loan_data.emp_length.fillna(loan_data.emp_length.mode()[0], inplace = True)
loan_data.emp_length.isna().sum()
```

Missing value percentage is very low in revol_util "0.13", so we can safely eliminate them

```python
loan_data.dropna(axis = 0, subset = ['revol_util'] , inplace = True)
loan_data.revol_util.isna().sum()
```

<h1>Standardizing the data</h1>

- "revol_util" has continous values. but it has "%" appended to it, So we need to standardize the data in this column
- "int_rate" is also has "%" appended to it, So we need to standardize the data in this column
- For this analysis let us asume {"emp_length" < 1 year) as 0 and 10+ years as 10 }

```python
loan_data.revol_util = pd.to_numeric(loan_data.revol_util.apply(lambda x : x.split('%')[0]))
```

```python
loan_data.int_rate = pd.to_numeric(loan_data.int_rate.apply(lambda x : x.split('%')[0]))
```

```python
loan_data.emp_length = pd.to_numeric(loan_data.emp_length.apply(lambda x: 0 if "<" in x else (x.split('+')[0] if "+" in x else x.split()[0])))
```
